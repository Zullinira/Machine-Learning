# Data Visualization

```python
plt.style.use('seaborn')
datairis['Species'].hist();
```

Double Variables

```python
datairis.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm');
```

Disribution plot

```python
sns.distplot(datairis['SepalLengthCm'], bins=10);
```

# Box Plot

```python
sns.boxplot(data=datairis)
plt.xlabel('Features')
plt.ylabel('Length in Cm')
plt.show()
```

```python
sns.boxplot(data=datairis, x='Species', y='SepalLengthCm');
```

# Multivariable-Pair Plot

```python
sns.pairplot(data=datairis,hue='Species');
```

# Show Correlation

```python
sns.heatmap(data=datairis.corr(), annot=True)
```

```python
sns.violinplot(data=datairis);
```
