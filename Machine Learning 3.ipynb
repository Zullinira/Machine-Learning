# Handling Categorical Value - One Hot Encoder

Fungsi get_dummies digunakan untuk mengubah variabel kategorikal menjadi variabel numerikal dengan melakukan proses One-Hot-Encode terhadap variabel kategorikal.
One-Hot-Encode adalah proses untuk membuat kolom baru dari variabel kategorikal kita di mana setiap kategori menjadi kolom baru dengan nilai 0 atau 1

```python
genderdummies=pd.get_dummies(data['Gender'], prefix='Gender')
genderdummies.head()
```

#### menggabungkan dataframe

```python
datagabung=pd.concat([data,genderdummies], axis=1)
```

# Handling Outlier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
#data set yang telah ada 
iris=datasets.load_iris()
x=iris.data
y=iris.target
columns=iris.feature_names

#Data Frame
datairis=pd.DataFrame(iris.data)
datairis.columns=columns
datairis.head()
```

```python
#cek outlier
```

```python
sns.boxplot(x=datairis['sepal length (cm)']);
```

```python
sns.boxplot(x=datairis['sepal width (cm)']);
```

terdapat beberapa outlier pada data, sehingga harus dikerluarkan data outlier dngan menggunakan interquartile

```python
q1= datairis.quantile(0.25)
q3= datairis.quantile(0.75)
```

```python
print(q1)
print(q3)
```

```python
#interquartile
iqr =q3-q1
print(iqr)
```

```python
#perlihatkan data outlier/ data yang berada diluar batas bawah dan atas
outlier=(datairis < q1-(1.5*iqr))|(datairis >(q3+(1.5*iqr)))
```

```python
datatanpaoutlier=datairis[~((datairis < q1-(1.5*iqr))|(datairis >(q3+(1.5*iqr))))]
```

```python
datatanpaoutlier.head()
```

```python
#cek ulang outlier
sns.boxplot(x=datatanpaoutlier['sepal width (cm)']);
```
